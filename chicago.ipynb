```python
import pandas as pd
import numpy as np

chicago_1=pd.read_csv('data/Chicago_Crimes_2001_to_2004.csv',error_bad_lines=False)
chicago_1.head()
```

```python
chicago_2=pd.read_csv('data/Chicago_Crimes_2005_to_2007.csv',error_bad_lines=False)
chicago_2.head()
```

```python
chicago_3=pd.read_csv('data/Chicago_Crimes_2008_to_2011.csv',error_bad_lines=False)
chicago_3.head()
```

```python
chicago_4=pd.read_csv('data/Chicago_Crimes_2012_to_2017.csv',error_bad_lines=False)
chicago_4.head()
```

```python
chicago=pd.concat([chicago_1,chicago_2,chicago_3,chicago_4])
```

```python
chicago.head()
```

```python
chicago['Date']=pd.to_datetime(chicago['Date'])
```

```python
chicago['WeekDay']=chicago['Date'].dt.weekday
```

```python
chicago['Hour']=chicago['Date'].dt.hour
```

```python
chicago['Month']=chicago['Date'].dt.month
```

```python
chicago.dropna(axis=0,how='any',subset=['Case Number'],inplace=True)
```

```python
chicago.isnull().sum()/chicago.shape[0]
```

```python
chicago.isnull().sum()
```

```python
homicide=chicago[chicago['Primary Type']=='HOMICIDE']
```
